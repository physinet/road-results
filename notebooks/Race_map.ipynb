```python
import folium
from folium.plugins import FastMarkerCluster

import pandas as pd
from datetime import datetime
```

```python
df = pd.read_pickle('C:/data/results/df.pkl')
df = df[df['json_url'] != '']
df['weather'] = df['weather'].str.split('\n').str.get(-1)  # sometimes the weather column also got "Series: " information
df_coord = df.dropna(subset=['coord'])
```

```python
df_coord.sample(5)
```

Coordinates for each race were not standard on the website until 2019, so we only have exact coordinates for the more recent data

```python
df.groupby(df['date'].dt.year).count()
```

```python
df_coord.groupby(df_coord['date'].dt.year).count()
```

```python
m = folium.Map(location=[39.8283, -98.5795], zoom_start=4)

def add_marker(coord, name=None):
    folium.CircleMarker([coord[0], coord[1]], popup=None, opacity=1, radius=1).add_to(m)

df_coord[df_coord['date'].dt.year == 2019].apply(lambda row: add_marker(row['coord'], row['name']), axis=1)


m.save()
```
